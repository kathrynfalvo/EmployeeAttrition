```python
# iii. How would the correlations change if we normalized the data first?

# First, recreate the correlation matrix, turning the negative values positive.
cor_mtx = hr3_df.corr().abs()
print(cor_mtx)
```

```python
# Because the matrix is identical if cut diagonally, only the triangle of the matrix above values mapped by k 
# is used by np.triu. 

# np.triu requires a value for k, which was set as 1 to enable np.ones that maps specified values as 1.
# .shape returns the number of rows and columns of an array while .astype converts values to True/False.
# Boolean values are expressed as NaN in numerical matrices, as with the case below (i.e. correlation).

upper_triangle = cor_mtx.where(np.triu(np.ones(cor_mtx.shape), 
                                          k=1).astype(bool))
print(upper_triangle)
```

```python
# Setup a list of columns to drop by iteratively assessing each column's correlation.

drop_these = [column for column in upper_triangle.columns if any(upper_triangle[column] > 0.7)]

print("Dataset features that can be dropped due to strong correlations:")
print(drop_these)

# Above 0.7 is considered a relatively high hcorrelation 
```

```python
from sklearn import preprocessing
norm_hr = preprocessing.normalize(hr3_df)
hr4_df = pd.DataFrame(norm_hr, columns=hr3_df.columns)
hr4_df.head()

```

```python
# This data then needed to be normalized to increase accuracy due to differences in scale causing reduced 
# pairwise correlations. 

norm_corr = hr4_df.corr()
plt.figure(figsize = (18,15))
sns.heatmap(norm_corr, xticklabels=norm_corr.columns, yticklabels=norm_corr.columns, 
            cmap="Blues", annot=True, fmt=".2f", cbar=True)
```

```python
# f. Apply of PCA for any two variables

import numpy as np
from sklearn.decomposition import PCA

pcs = PCA(n_components=2)
pcs.fit(hr4_df[['JobSatisfaction','WorkLifeBalance']])
pcs_summary = pd.DataFrame({'Standard Deviation' : np.sqrt(pcs.explained_variance_),
                           'Ratio' : pcs.explained_variance_ratio_,
                           'Cumulative Proportion' : np.cumsum(pcs.explained_variance_ratio_)})

pcs_summary = pcs_summary.transpose()
pcs_summary.columns = ['PC1', 'PC2']
pcs_summary.round(2)
```

```python
pcs_comp_df = pd.DataFrame(pcs.components_.transpose(), columns=['PC1','PC2'], index=['JobSatisfaction','WorkLifeBalance'])
pcs_comp_df
```
